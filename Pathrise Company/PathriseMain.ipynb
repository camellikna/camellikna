# Missing that should be Missing value

```python
# Rename to Missing Values
df['gender'].replace(np.nan, 'missing_gender', inplace=True)
df['work_authorization_status'].fillna('MISSING_WORK_AUTH', inplace = True)
df['employment_status '].replace(np.nan, 'missing_emply_status', inplace=True)
df['professional_experience'].replace(np.nan, 'missing_pro_exp', inplace=True)
df['length_of_job_search'].replace(np.nan, 'missing_job_searc', inplace=True)
df['highest_level_of_education'].replace(np.nan, 'missing_edu_lvl', inplace=True)
```

```python
df.info()
```

# def for checkin normalization

```python
# Run a normality test to determine if I should impute using the mean or median
def is_normal(data):

  _,p = stats.normaltest(data, nan_policy = 'omit')

  if p < 0.05:
        return False
  else:
        return True
```

# Number of Interviews



```python
is_normal(df['number_of_interviews'])
```

```python
interview_num_median=df['number_of_interviews'].median()
df['number_of_interviews'].replace(np.nan , interview_num_median, inplace = True)
```

```python
df.info()
```

# Race

```python
df['race'].replace(np.nan , df['race'].mode()[0], inplace=True)
```

# Cohort Tag

```python
df['cohort_tag'].replace(np.nan , df['cohort_tag'].mode()[0], inplace=True)
```

# Biggest chalenge in the search

```python
df['biggest_challenge_in_search'].replace(np.nan, df['biggest_challenge_in_search'].mode()[0],inplace=True)
```

```python
df.info()
```

##program_duration_days

```python
is_normal(df['program_duration_days'])
```

```python
categ_vars = ['gender', 'race', 'pathrise_status', 'employment_status ',
              'highest_level_of_education', 'length_of_job_search',
              'professional_experience', 'work_authorization_status',
              'primary_track', 'cohort_tag', 'placed']

quant_vars = ['number_of_interviews', 'number_of_applications']
```

```python
impute_data = pd.get_dummies(df[categ_vars])
```

```python
impute_data
```

```python
impute_data[quant_vars] = df[quant_vars]
```

```python
impute_data['program_duration_days'] = df['program_duration_days']
```

```python
from sklearn.linear_model import LinearRegression
```

Predict Data for program duration days

```python
train =impute_data[impute_data['program_duration_days'].notnull()]
test =impute_data[impute_data['program_duration_days'].isnull()]
```

```python
x_train = train.drop(columns=['program_duration_days'])
y_train = train['program_duration_days']
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
x_test = test.drop(columns=['program_duration_days'])
```

```python
lr.predict(x_test)[:10]
```

```python
from sklearn.linear_model import Lasso
clf = Lasso(alpha = 0.1)
clf.fit(x_train, y_train)


clf.predict(x_test)[0:10]
```

```python
df.loc[test.index.values, "program_duration_days"] = clf.predict(x_test)
```
